```python
import pandas as pd
import os
import minsearch
from dotenv import load_dotenv
import numpy
import json
```

```python
with open('../data/radiologydata_cleaned.json', 'rt') as f_in:
    documents = json.load(f_in)
```

```python
index = minsearch.Index(
    text_fields=['uid', 'article_date', 'article_type', 'specialty', 'publisher',
       'comparison', 'indication', 'findings', 'impression', 'mesh_terms',
       'img_id', 'figure_id', 'caption', 'url', 'img_modality', 'cedd',
       'colorlayout', 'edgehistogram', 'fcth', 'semanticcontext60'],
    keyword_fields=[]
)
```

```python
index.fit(documents)
```

```python
query = 'List all imaging done for trauma cases'
```

```python
index.search(query, num_results=10)
```
